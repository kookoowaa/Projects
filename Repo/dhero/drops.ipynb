```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import re
import json
```

```python
tmp = open('sample.html', 'r', encoding='utf-8').read()
```

```python
tmp = bs(tmp, 'html.parser')
```

```python
[i.get_text() for i  in tmp.find_all('label')]
```

```python
badge_info = {}
```

```python
def badge_loc(badge, loc):
    outcome = {'badge':[], 'loc':[]}
    for i in range(len(loc)):
        match_loc = re.compile('캠페인')
        if bool(match_loc.match(loc[i])):
            outcome['badge'].append(badge)
            outcome['loc'].append(loc[i])
    return outcome
```

```python
for badge in tmp.find_all('tr'):
    badge_tmp = badge.get_text()
    try:
        badge_tmp = re.split(r'\n\n\n', badge_tmp)
    
        target = re.sub(r'\n','',badge_tmp[1])
        if bool(re.match('캠페인', badge_tmp[3])):
            loc = re.split(r'\n', badge_tmp[3])

            loc_info = {}
            for i in loc:
                loc_info['{}'.format(i)] = 1

            badge_info['{}'.format(target)]= loc_info
        
        
    except:
        pass
    
    
    
```

```python
badge_table = pd.DataFrame(badge_info).T
badge_table
```
