```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
tmp = pd.read_csv('mb(nov).csv', engine = 'python')
tmp.head()
```

```python
tmp = tmp[tmp.HFB!=7]
```

```python
test_set = tmp.groupby(['Order Number', 'Article No'])['Quantity'].count().unstack().reset_index().fillna(0).set_index('Order Number')
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
```

```python
test_set = test_set.applymap(encode_units)
```

```python
test_set
```

```python
test_result = apriori(test_set, min_support = 0.005, use_colnames=True)
test_result
```

```python
rules = association_rules(test_result, metric = 'lift', min_threshold=1)
rules
```

```python
rules.to_csv('out.csv')
```
