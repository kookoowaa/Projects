### impossible values

- negative figures is impossible on daily summed POS data
- Order counts may have integrity issues (i.e. # of beer & soju order exceeding # of beer order)

```python
# to test if any value in row contains negative value
def detect_integrity(row):
    
    ln = len(row)
    test = 0
    
    for i in range(ln):
        test += (row[i] < 0)
    # test each value and add count if wrong
    
    result = False
    if test != 0:
        result = True
    # if count is not 0, return False
    
    return result
```

### Regression: Bagging + Ridge

```python
def bootsTrap(data, size=0.7, seed=None):
    if seed==None:
        pass
    else:
        np.random.seed(seed)
    tmpIdx = np.random.choice(data.index, size = int(size*len(data)))
    tmpIdx = pd.DataFrame(tmpIdx, columns = ['idx'])
    tmpdf = pd.merge(tmpIdx, data, left_on='idx', right_index=True)
    return tmpdf.drop('idx', axis=1).reset_index(drop=True)

# sampling with replacement
```

```python
true_y = testSetReg.beerU
true_X = testSetReg.iloc[:,4:].drop('beerU', axis=1)
```

```python
tmp = bootsTrap(testSetReg, seed=1)
y = tmp.beerU
X = tmp.iloc[:,4:].drop('beerU', axis=1)
model = Ridge(alpha=0.5, normalize=True)
model.fit(X, y)
pd.DataFrame(model.coef_, index = X.columns, columns = ['coef']).T

# sample Ridge regression over 1 subsample
```

```python
R2 = model.score(true_X, true_y)
MSE = mean_squared_error(model.predict(true_X), true_y)
print('''
The model above returns following scores to the true values
R2: {R2}
MSE: {MSE}
'''.format(R2=round(R2,4), MSE=round(MSE,4)))
```

```python
# Run 100 Ridge regression for aggregating for a robust model
def baggedRidge(data, r=100, alphas = [0.001, 0.005, 0.01, 0.05]):
    metrics = []
    for r in range(r):
        tmp = bootsTrap(data, seed=r)
        y = tmp.beerU
        X = tmp.iloc[:,4:].drop('beerU', axis=1)

        model = RidgeCV(alphas=alphas, normalize=True)
        model.fit(X, y)

        tmp = []
        tmp.append(model.intercept_)
        tmp.extend(model.coef_)
        R2 = model.score(true_X, true_y)
        tmp.append(R2)
        MSE = mean_squared_error(model.predict(true_X), true_y)
        tmp.append(MSE)
        metrics.append(tmp)
        
    varNames = ['intercept']
    varNames.extend(X.columns)
    varNames.append('R2')
    varNames.append('MSE')
    
    output = pd.DataFrame(metrics, columns = varNames)
        
        
    return model, output
```

```python
model, metrics = baggedRidge(testSetReg,100)
```

```python
print(metrics.shape)
metrics.head()
```

```python
metrics[['R2', 'MSE']].describe()
```

- The bootstraped models have around R2 score of 0.83, which means the models can roughly explain 83% of variance.
- MSE is around at 0.16, both of which are acceptable for the purpose of understanding the effect of each variables on "beer demand"  

*the scores are on full dataset (includes oob observations)*

```python
def demandCall(metrics, period='', d=2, X=true_X, y=true_y):
    effects = metrics[['sojuU', 'spiritsU', 'foodR', 'nonAlcoholR']].mean()
    tmp = model
    tmp.coef_ = metrics.iloc[:,1:-2].mean()
    tmp.intercept_ = metrics['intercept'].mean()
    R2 = tmp.score(X, y)
    MSE = mean_squared_error(y, tmp.predict(X))
    print('''
    From data{p}
    % change in Soju Units would affect {sojuU}% in beer units to be sold,
    % change in Spirits Units would affect {spiritsU}% in beer units to be sold,
    % change in Food Revenue would affect {foodR}% in beer units to be sold,
    % change in Non Alcoholic Revenue  would affect {nonAlcoholR}% in beer units to be sold.
    
    The model has R2 score of {r2}, and MSE score of {mse}.
    '''.format(p = period,
              sojuU = round(effects['sojuU'],d),
              spiritsU = round(effects['spiritsU'],d),
              foodR = round(effects['foodR'],d),
              nonAlcoholR = round(effects['nonAlcoholR'],d),
              r2 = round(R2,4), mse = round(MSE,4)))
```

```python
metrics[['sojuU', 'spiritsU', 'foodR', 'nonAlcoholR']].describe()
```

```python
demandCall(metrics)
```

- Aggregating the bootstrapped models for a robust bagged model, it returns above result, R2 at 0.841, MSE at 0.158.
- Soju units seem to have independent relationship with beer units.
- Yet the coefficient of Soju Units vary from -0.035 to 0.015, which may need more study. (perhaps looking at the effects by period)

## 3. Additional analysis that could provide insight to activate beer category

- As previously looked for, the characterisitcs of bar are diverse, looking at their sales ratio.
- Grouping the bars(kmeans) for above analysis could help us find insights and tailored measures that can increase beer sales.


```python
barChar = testSet.iloc[:,:13].groupby('barId', as_index=False).mean()
barChar.head()
```

```python
inertias = []

for k in range(10):
    model = KMeans(n_clusters=k+1)
    model.fit(barChar.iloc[:,1:])
    inertias.append(model.inertia_)
    
plt.plot(range(1,11), inertias, '-o')
    
```

- It seems appropriate to cluster bars in 4 groups.

```python
model = KMeans(n_clusters=4)
model.fit(barChar.iloc[:,1:])
barChar['cluster'] = model.labels_
barChar.head()
```

```python
pca = PCA(n_components=2, random_state=0)
tmp = pd.DataFrame(pca.fit_transform(barChar.iloc[:,1:-1]), columns = ['pca1', 'pca2'])
tmp = pd.concat([tmp, barChar.cluster], axis=1)
sns.scatterplot(data=tmp, x='pca1', y='pca2', hue='cluster', palette='Set2')
plt.title("Visualizing clusters in PCA dimention")
```

```python
barChar.groupby('cluster')['barId'].count()
```

```python
barChar.groupby('cluster').mean()
```

- It seems the clusters are well splited, as each cluster demonstrates unique characteristics:
>- cluster0: has 87 bars(the most) in cluster with low sales volume of beer.
>- cluster1: has 11 bars in cluster with fairly high sales volume of everything.
>- cluster2: has 48 bars, and proportion of beer sales is relatively high.
>- cluster3: has only 1 bar, and very low alcohol sales compared to food and nonAlcohol, which I assume to be a normal restaurant
- Now run regression for all four clusters

```python
# Run 100 Ridge regression for aggregating for a robust model per month

coef = {}
for c in range(4):
    coef['cluster{}'.format(c)]=[]
    barId = barChar[barChar.cluster==c].barId.unique()
    for r in range(100):
        tmp = bootsTrap(testSetReg[testSetReg.barId.isin(barId)], seed=r)
        y = tmp.beerU
        X = tmp.iloc[:,4:].drop('beerU', axis=1)
        model = Ridge(alpha=0.5, normalize=True)
        model.fit(X, y)
        coef['cluster{}'.format(c)].append(model.coef_)
```

```python
tmp = pd.DataFrame([pd.DataFrame(coef['cluster0']).mean(), pd.DataFrame(coef['cluster1']).mean(),
                   pd.DataFrame(coef['cluster2']).mean(), pd.DataFrame(coef['cluster3']).mean()],
                   index = ['cluster{}'.format(i) for i in range(4)])
tmp.columns = X.columns
```
