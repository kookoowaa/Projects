```python
tmp = dict(rd.iloc[0,:7])
tmp['stats'] = dict(rd.iloc[0,7:17])
tmp['skills'] = dict(rd.iloc[0,17:])
```

```python
tmp
```

```python
knights = []

for i in range(len(rd)):
    tmp = dict(rd.iloc[i, :7])
    tmp['stats'] = dict(rd.iloc[i, 7:17])
    tmp['skills'] = dict(rd.iloc[i, 17:])
    knights.append(tmp)
```

```python
knights = {'knights': knights}
```

```python
type(knights['knights'][0]['star'])
```

```python
for knight in knights['knights']:
    knight['star'] = int(knight['star'])
    knight['salary'] = int(knight['salary'])
    knight['stats']['dmg'] = int(knight['stats']['dmg'])
    knight['stats']['att_speed'] = float(knight['stats']['att_speed'])
    knight['stats']['run_speed'] = float(knight['stats']['run_speed'])
    knight['stats']['crit_rate'] = float(knight['stats']['crit_rate'])
    knight['stats']['crit_dmg'] = float(knight['stats']['crit_dmg'])
    knight['stats']['hp'] = int(knight['stats']['hp'])
    knight['stats']['phy_def'] = float(knight['stats']['phy_def'])
    knight['stats']['mag_def'] = float(knight['stats']['mag_def'])
    knight['stats']['evasion'] = float(knight['stats']['evasion'])
    knight['stats']['att_range'] = int(knight['stats']['att_range'])
    knight['skills']['skill1_rate'] = float(knight['skills']['skill1_rate'])
    knight['skills']['skill2_rate'] = float(knight['skills']['skill2_rate'])
    knight['skills']['skill3_rate'] = float(knight['skills']['skill3_rate'])
```

```python
a = json.dumps(knights, ensure_ascii=False)
```

```python
with open('char.json', 'w', encoding = 'utf-8') as f:
    f.write(a)
    f.close()
```
