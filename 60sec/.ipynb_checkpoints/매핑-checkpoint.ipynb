```python
import re

class numberConversion:
    
    def __init__(self, *args):
        alphabets = list('abcdefghijklmnopqrstuvwxyz')
        k = []
        for char in alphabets:
            for char2 in alphabets:
                k.append(char+char2)
        self.k = alphabets + k
        self.n = list(*args)
        
    def nToChar(self, number):
        tmp = number
        i = 0
        
        while tmp >1000:
            tmp = round(tmp/1000, 2)
            i += 1
        
        return str(tmp)+self.k[i-1]
    
    def charToN(self, char):
        span = re.search(pattern='[a-z]', string = char).span()[0]
        
        num = float(char[:span])
        char = char[span:]
        digit = self.k.index(char)
        
        return int(num*1000**(digit+1))
```
